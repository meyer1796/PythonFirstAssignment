```python
import pandas as pd
```

```python
df = pd.read_csv('regrex1.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
df['x'].head()
```

```python
df['y'].head()
```

```python
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=np.array(df['x']).reshape((-1,1))
```

```python
y=np.array(df['y'])
```

```python
model = LinearRegression()
```

```python
md=model.fit(x,y)
```

```python
md.coef_
```

```python
md.intercept_
```

```python
yp = md.predict(x)
```

```python
plt.plot(x, yp)
```

```python
plt.scatter(x,y)
plt.plot(x,yp)
plt.show()
```
